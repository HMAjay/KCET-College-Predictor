# KCET College Predictor — Data Exploration
Explore the master dataset, check distributions, and validate model quality.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from config import MASTER_CSV, MODEL_PATH
sns.set_theme(style='whitegrid')
```

## 1. Load Master Dataset

```python
df = pd.read_csv(MASTER_CSV)
df['Cutoff_Rank'] = pd.to_numeric(df['Cutoff_Rank'], errors='coerce')
print(df.shape)
df.head(10)
```

## 2. Basic Stats

```python
print('Colleges :', df['College_Code'].nunique())
print('Branches :', df['Branch'].nunique())
print('Categories:', df['Category'].nunique())
print('Years    :', sorted(df['Year'].dropna().unique()))
df.describe()
```

## 3. Cutoff Rank Distribution

```python
fig, ax = plt.subplots(figsize=(12,5))
df['Cutoff_Rank'].clip(upper=100000).hist(bins=80, ax=ax, color='steelblue', edgecolor='white')
ax.set_title('Cutoff Rank Distribution (clipped at 1,00,000)')
ax.set_xlabel('Cutoff Rank')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

## 4. Top Branches by Row Count

```python
top_branches = df['Branch'].value_counts().head(15)
top_branches.plot(kind='barh', figsize=(10,5), color='coral')
plt.title('Top 15 Branches')
plt.tight_layout()
plt.show()
```

## 5. Category Distribution

```python
df['Category'].value_counts().plot(kind='bar', figsize=(14,4), color='mediumseagreen')
plt.title('Category Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Year-wise Cutoff Trend (GM, CSE)

```python
cse_gm = df[
    (df['Branch'].str.contains('Computer Science', case=False, na=False)) &
    (df['Category'] == 'GM')
].groupby('Year')['Cutoff_Rank'].median()

cse_gm.plot(marker='o', figsize=(9,4), color='royalblue')
plt.title('Median CSE GM Cutoff by Year')
plt.ylabel('Median Cutoff Rank')
plt.tight_layout()
plt.show()
```
